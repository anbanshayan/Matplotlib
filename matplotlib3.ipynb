```python
runs_per_ball = [4,1,1,0,1,6,0,4,1,0,1,5,0,1,1,1,0,2,2,1,0,0,1,0,2,0,4,4,1,1,1,2,0,0,6,0,0,1,0,1,0,1,1,6,1,0,6,1,1,1,0,1,0,1,4,0,1,0,1,1,6,0,1,0,0,0,1,0,2,0,2,6,0,0,0,1,1,1,2,0,2,0,1,1,4,1,6,1,0,1,1,1,6,2,2,1,6,0,0,6,0,1,1,2,0,2,2,1,0,6,0,0,4,2,1,0,6,6,4,6,6]

# Calculate cumulative runs
cumulative_runs = np.cumsum(runs_per_ball)

# Find number of complete overs
num_overs = len(runs_per_ball) // 6

# Get the cumulative runs at the end of each over
runs_at_end_of_overs = [cumulative_runs[(i+1)*6 - 1] for i in range(num_overs)]

# Create over numbers starting from 1
over_numbers = list(range(1, num_overs + 1))

# Plot
plt.figure(figsize=(10, 5))
plt.plot(over_numbers, runs_at_end_of_overs, marker='o', linestyle='-', color='red')
plt.title('Cumulative Runs of PBKS at the End of Each Over in IPL2025')
plt.xlabel('Over Number')
plt.ylabel('Cumulative Runs')
plt.grid(True)
plt.xticks(over_numbers)
plt.tight_layout()
plt.savefig("pbks_innings.jpg")
plt.show()
```
