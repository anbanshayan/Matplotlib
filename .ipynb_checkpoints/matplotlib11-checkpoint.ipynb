```python
Product = ['Apple','Banana','Orange']
Region_A = [30,20,10]
Region_B = [25,15,20]
Region_C = [10,30,15]

Region_C_start = [Region_A[i]+Region_B[i] for i in range(len(Region_A))]

plt.bar(Product,Region_A,width=0.4,color="Red",label='Region A')
plt.bar(Product,Region_B,bottom = Region_A,width=0.4,color="Yellow",label='Region B')
plt.bar(Product,Region_C,bottom = Region_C_start,width=0.4,color="orange",label='Region C')
plt.legend()
plt.xlabel("Fruits")
plt.ylabel("Counts on Yield")
plt.title("Sample Complete Stack Bar Plot")
plt.show()
```
