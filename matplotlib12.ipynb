```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
student_performance = ['Excellent','Good','Average','Poor']
student_count = [39,20,11,32]
plt.title("Performance of the students")
plt.pie(student_count,labels=student_performance,startangle=0,colors=['blue','green','yellow','red'])
plt.legend()
plt.show()
```

**Sample Pie Chart**

```python
product_categories = ['Fruits','Vegetables','Dairy','Snacks','Beverages']
product_sales_units = [150,120,90,80,60]
plt.pie(product_sales_units,labels=product_categories)
plt.legend()
plt.title("Supermarket Product Categories")
plt.show()
```
