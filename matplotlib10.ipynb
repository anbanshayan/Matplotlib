**Scatterplot**

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0,12,1)
y1 = x+1
y2 = 2*x
plt.scatter(x,y1,color="orange",marker="s",linewidth=10,alpha=0.5)
plt.scatter(x,y2,color="blue",linewidth=10,alpha=0.5)
plt.show()
```

**Color maps representing colors in terms of integers**

```python
x = [10,20,30,40,50]
y=[200,300,400,500,400]
colors = ["red","orange","blue","black","green"]
sizes = [100,50,300,250,150]
plt.scatter(x,y,c=colors,s=sizes)
colors = ["red","pink","black","blue","yellow"]
plt.title("Scatterplot welcomes you!")
plt.show()
```
