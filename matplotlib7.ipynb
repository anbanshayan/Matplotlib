```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0,12,2)
y = x+3
```

```python
a = np.arange(0,10,2)
b = a+1
```

```python
plt.plot(x,y)
plt.plot(a,b)
```

```python
c = np.arange(0,20,2)
d = c+5
e = c+7
f = c+8
g = c**2
```

```python
plt.subplot(2,2,1)
plt.plot(c,d)
plt.subplot(2,2,2)
plt.plot(c,e)
plt.subplot(2,2,3)
plt.plot(c,f)
plt.subplot(2,2,4)
plt.plot(c,g)
plt.suptitle("Sample Graphs")
plt.show()
```
