**Legend function**

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0,12,3)
```

```python
y1 = x+2
y2 = x**3
```

```python
plt.plot(x,y1,label="Y1")
plt.plot(x,y2,label="Y2")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph with legend function")
plt.grid(axis="both",ls="dotted",lw=2,color="black")
plt.legend(loc=1,facecolor="lightblue",shadow=True,fancybox=True)
plt.show()
```
