**Plot the following data using plot():**

```python
from matplotlib import pyplot as plt
import numpy as np
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
```

**Plot a graph of y = x² for x values ranging from 0 to 10 using a simple line plot.**

```python
x = np.arange(0,20,2)
y = x**2
plt.plot(x,y)
```

**Plot the points x = [1, 2, 3, 4] and y = [10, 20, 25, 30] with a red circle ('ro') marker on each point.**

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y,marker="o",c="r")
```

**Plot two lines on the same graph:
y = 2x as a dashed green line
y = x + 5 as a solid blue line**

```python
x = np.arange(0,10,2)
y = x*2
z = x+5
plt.plot(x,y,ls="--",color="green")
plt.plot(x,z,ls="-",color="blue")
plt.title("Two graphs oo")
plt.show()
```

**Add labels to the X and Y axes with the font family 'serif' and size 14, and give the plot a title in bold and size 16.**

```python
x = np.arange(0,10,2)
y = x+3
f1={"family":"serif","size":14}
f2={"family":"serif","size":16,"weight":"bold"}
plt.xlabel("X axis",fontdict=f1)
plt.ylabel("Y axis",fontdict=f1)
plt.title("The graph",fontdict=f2)
plt.plot(x,y)
plt.show()
```
