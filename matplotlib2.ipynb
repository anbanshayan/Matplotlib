```python
from matplotlib import pyplot as plt
import numpy as np
x = np.array([i for i in range(-10,11)])
y = x**2
plt.plot(x,y,marker="o",ms="9")
plt.title("X squared Graph")
plt.xlabel("numbers range from minus to plus")
plt.ylabel("their positive results")
plt.show()
```

```python
a = np.array([i for i in range(0,9)])
b = a**2+(5*a)+6
plt.plot(a,b,marker="p",ms="10")
plt.title("Graph of a**2+5a+6")
plt.xlabel("values range from 0 to 9")
plt.ylabel("Equation Values")
plt.show()
```

```python
c = np.arange(0,11)
d = c**3
plt.title("X cubed graph")
plt.xlabel("X Axis")
plt.ylabel("Y axis")
plt.plot(c,d,ls="dashed")
plt.show()
```
