**Font Properties**

```python
from matplotlib import pyplot as plt
import numpy as np
```

**Two forn dictionaries are defined here**

**ls = > line size ,
lw = > dotted , c=> color**

```python
x = np.arange(0,12,2)
y = x+4
z = -x+5
f1 = {'family':'sans-serif','color':'blue','size':17}
f2 = {'family':'monospace','color':'violet','size':14}
plt.title("Graph Series with font custom",fontdict=f2)
plt.xlabel("X axis",fontdict=f1)
plt.ylabel("Y axis",fontdict=f1)
plt.plot(x,y,ls="dotted",lw="12",c='k',label="x+4")
plt.plot(x,z,ls="dotted",lw="12",c='red',label="-x+5")
plt.legend()
plt.show()
```
