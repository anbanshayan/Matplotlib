**Font Properties**

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0,12,2)
y = x+4
f1 = {'family':'sans-serif','color':'blue','size':17}
plt.title("Graph Series with font custom",fontdict=f1)
plt.xlabel("X axis",fontdict=f1)
plt.ylabel("Y axis",fontdict=f1)
plt.plot(x,y)
plt.show()
```
